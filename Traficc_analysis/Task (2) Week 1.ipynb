## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd
df= pd.read_csv('Task (2) Dataset.csv')
df
```

### Identify and analyze patterns in the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['Datetime'] = pd.to_datetime(df['Datetime'])
print(df.describe(include='all'))



plt.figure(figsize=(12, 6))
sns.lineplot(x='Datetime', y='TrafficIndexLive', hue='City', data=df)
plt.title('Traffic Index Live Over Time')
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df['Date'] = df['Datetime'].dt.date
plt.figure(figsize=(12, 6))
sns.histplot(df['TrafficIndexLive'], bins=20, kde=True)
plt.title('Distribution of Traffic Index Live')
plt.xlabel('Traffic Index Live')
plt.ylabel('Frequency')
plt.show()
```
