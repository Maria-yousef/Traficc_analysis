## Data Preprocessing
Do what you think you need such as:
* Remove the outliers
* Impute missing data
* Scale the data
* Reduce dimentions using PCA
* Implement One-Hot Encoding for nominal categorical variables.

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns
def remove_outliers_iqr(df, columns):
    df_cleaned = df.copy()
    for col in columns:
        Q1 = df_cleaned[col].quantile(0.25)
        Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        df_cleaned = df_cleaned[(df_cleaned[col] >= (Q1 - 1.5 * IQR)) & (df_cleaned[col] <= (Q3 + 1.5 * IQR))]
    return df_cleaned

data_cleaned = remove_outliers_iqr(df, numeric_columns)
```

```python
#Impute missing data

imputer = SimpleImputer(strategy='median')
data_cleaned[numeric_columns] = imputer.fit_transform(data_cleaned[numeric_columns])
```

```python
#Scale the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df[numeric_columns])
```

```python
#Reduce dimentions using PCA

pca = PCA(n_components=2)
principal_components = pca.fit_transform(data_cleaned[numeric_columns])
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
data_cleaned_pca = pd.concat([data_cleaned, pca_df], axis=1)
```

* Find the `n_clusters` parameter using the elbow method.
* Train the model.

```python
def calculate_elbow_and_draw_it(data):
    wcss = []
    for i in range(1, 11):
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)
    
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, 11), wcss, marker='o')
    plt.title('The Elbow Method')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
    plt.grid(True)
    plt.show()
calculate_elbow_and_draw_it(df_encoded.values)
```
